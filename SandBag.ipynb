```
from imageai.Detection.Custom import CustomObjectDetection

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("/content/sandbag/models/detection_model-ex-001--loss-0145.753.h5") 
detector.setJsonPath("/content/sandbag/json/detection_config.json")
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image="/content/sandbag/Royal-Oak-Charcoal-Briquettes -1.jpg", output_image_path="/content/sandbag/output.jpg")
for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
```
